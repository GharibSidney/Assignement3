```python
new_param_grid_decision_tree = {
    'max_depth': [None, 5, 10],
}

new_param_grid_random_forest = {
    'n_estimators': [5, 10, 30],
}

new_param_grid_svm = {
    'kernel': {'linear', 'poly', 'rbf'}
}
from sklearn.metrics import f1_score, accuracy_score
```

```python
import matplotlib.pyplot as plt
def plot_accuracy_hyp(hyperparams, accuracy_array, model_name, hyper_name, xlabel):
    plt.figure(figsize=(10, 6))
    plt.plot(hyperparams, accuracy_array, color='blue', marker='o')
    plt.title('ACcuracy for {} with different {} values'.format(model_name, hyper_name))
    plt.xlabel(xlabel)
    plt.ylabel('Accuracy')
    plt.show()
```

```python
hyperparameter_array = []
accuracy_array = []
for depth in new_param_grid_decision_tree['max_depth']:
    hyperparameter_array.append(depth)
    cls = DecisionTreeClassifier(random_state=0, max_depth=depth)
    cls.fit(X_train_scaled, y_train)
    y_train_pred = cls.predict(X_train_scaled)
    y_test_pred = cls.predict(X_test_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred) 
    accuracy_array.append(train_accuracy)
```

```python
plot_accuracy_hyp(hyperparameter_array, accuracy_array, 'Decision Tree', 'max_depth', 'max_depth')
```

```python
hyperparameter_array = []
accuracy_array = []
for estimator in new_param_grid_random_forest['n_estimators']:
    cls = RandomForestClassifier(random_state=0, n_estimators=estimator )
    cls.fit(X_train_scaled, y_train.values.ravel())
    y_train_pred = cls.predict(X_train_scaled)
    y_test_pred = cls.predict(X_test_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred) 
    hyperparameter_array.append(estimator)
    accuracy_array.append(train_accuracy)
plot_accuracy_hyp(hyperparameter_array, accuracy_array, 'Random Forest', 'n_estimators', 'n_estimators')
```

```python
hyperparameter_array = []
accuracy_array = []
for method in new_param_grid_svm['kernel']:
    cls = SVC(random_state=0, kernel=method )
    cls.fit(X_train_scaled, y_train.values.ravel())
    y_train_pred = cls.predict(X_train_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    hyperparameter_array.append(method)
    accuracy_array.append(train_accuracy) 

plot_accuracy_hyp(hyperparameter_array, accuracy_array, 'SVM', 'kernel', 'kernel')
```
